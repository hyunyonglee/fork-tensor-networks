```julia
include("../src/ForkTensorNetworks.jl")
using ITensors, ITensorMPS
using .ForkTensorNetworks
```

```julia
# Define the size of the lattice

Lx = 4
Ly = 5
χˣ = 4
χʸ = 6

# Define the physical index
phys_idx = siteinds("S=1/2", Lx * Ly);

# Create a random state in the Fork Tensor Network State representation
ψ = ForkTensorNetworkState(Lx, Ly, phys_idx; χˣ, χʸ);
```

```julia
# Plotting the network of which ψ is not in canonical form
plot_network(ψ)
```

```julia
# Norm of the random ftns
@show norm_ftn(ψ);

# Normalize the ftns 
normalize_ftn!(ψ); 
@show norm_ftn(ψ);
```

```julia
## Note: Normalization brings the ftns to the canonical form of which the canonical center is at (1, 1)
# Plotting the network
plot_network(ψ)
```

```julia
# Note: The red site is the canonical center, and the arrow indicates the direction of the isometries
# Moving the center to the right
canonical_center_move!(ψ, "right")
plot_network(ψ)
```

```julia
# Moving the center to the left
canonical_center_move!(ψ, "left")
plot_network(ψ)
```

```julia
# Moving the center to the down
canonical_center_move!(ψ, "down")
plot_network(ψ)
```

```julia
# Moving the center to the up
canonical_center_move!(ψ, "up")
plot_network(ψ)
```

```julia
# Note: One can also move the canonical center to a specific site
# Moving the center
canonical_form!(ψ, 3, 3) # Move the center to the site (x, y) = (3, 3)
plot_network(ψ)
```
